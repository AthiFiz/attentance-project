```python
import insightface
```

```python
from insightface.app import FaceAnalysis
```

```python
model_pack_name = 'buffalo_sc'
app = FaceAnalysis(name=model_pack_name, 
                   root="../app/insightfce_model/buffalo_sc/", 
                   providers=["CPUExecutionProvider"])

app.prepare(ctx_id=0, det_size=(640, 640))
```

```python
img = cv2.imread("../people.jfif")

cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
results = app.get(img)
```

```python
img_copy = img.copy()
for res in results:

    bbox = res["bbox"].astype(int)
    x1,y1,x2,y2 = bbox
    cv2.rectangle(img_copy, (x1,y1), (x2, y2), (0,255,0), 1)

    kps = res["kps"].astype(int)
    for k1, k2 in kps:
        cv2.circle(img_copy,(k1, k2), 1, (255,0,0), -1)
    
    score = res["det_score"]
    score = f"Score: {round(score*100)}%"
    cv2.putText(img_copy, score, (x1,y1), cv2.FONT_HERSHEY_COMPLEX_SMALL, fontScale=0.5, color=(255,255,255), thickness=1)
```

```python
cv2.imshow("img", img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
